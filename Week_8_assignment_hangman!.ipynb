```
# Hangman game !

print("Welcome to Hangman!")
print("You will be allowed 6 wrong guesses total.")

# bank of holding things
wrong_guesses = 0
blank_list = []

def choose_word():
  return "balloon"

# function result -> variable to use -> length
my_string = choose_word()
length_string = len(my_string)

# my_string is the variable for the WORD
# length_string is variable for the LENGTH of the word


def stickdraw(wrong_guesses):
  if wrong_guesses == 1:
    print("  (_)  ")
  
  if wrong_guesses == 2:
    print("  (_)  ")
    print("   |   ")
    print("   |   ")
  
  if wrong_guesses == 3:
    print("  (_)  ")
    print("  /|   ")
    print(" / |   ")
  
  if wrong_guesses == 4:
    print("  (_)  ")
    print("  /|\  ")
    print(" / | \ ")
  
  if wrong_guesses == 5:
    print("  (_)  ")
    print("  /|\  ")
    print(" / | \ ")
    print("  /    ")
    print(" /     ")
  
  if wrong_guesses == 6:
    print("  (_)  ")
    print("  /|\  ")
    print(" / | \ ")
    print("  / \  ")
    print(" /   \ ") 



# show up as blanks
for blanks in range(length_string):
  blank_list.append("_")

# ask for input
while "_" in blank_list:
  useranswer = input("Please guess a letter: ")

  for blanks in range(length_string):
    if useranswer.lower() == my_string[blanks]:
      blank_list[blanks] = useranswer

  if not useranswer in my_string[blanks]:
    wrong_guesses = wrong_guesses + 1
    print(stickdraw(wrong_guesses))

    if wrong_guesses == 6:
      break
      
    print("There are a total of " + str(wrong_guesses) + " wrong guesses.")
  print(blank_list)

print("You lost! :( The word was '" + my_string + "'.")





```
